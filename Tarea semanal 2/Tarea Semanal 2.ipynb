```python
from scipy import signal
import matplotlib.pyplot as plt
from splane import bodePlot, pzmap
from IPython.display import Image
import math
```

```python
w_0 = 1
R_2 = 10
R_3 = 1
Q = R_2/R_3
Q_butterworth = 1/math.sqrt(2)
K = - (R_3 / R_2)
```

```python
numerador = [1]
```

```python
denominador = [1, 1/Q, 1]
denominador_butterworth = [1, 1/Q_butterworth, 1]
```

```python
sys = signal.TransferFunction(numerador, denominador)
sys2 = signal.TransferFunction(numerador, denominador_butterworth)
w, mag, phase = signal.bode(sys)
w2, mag2, phase2 = signal.bode(sys2)
```

```python
plt.figure()
plt.semilogx(w, mag) # Bode magnitude plot
plt.semilogx(w2+0, mag2) # Bode magnitude plot
```

La diferencia es tan sútil que los gráficos parecen superpuestos.

```python
plt.figure()
plt.semilogx(w, phase) # Bode phase plot
plt.semilogx(w2, phase2) # Bode phase plot
```

```python
pzmap(sys) # Diagrama de polos y ceros. Función con efectos colaterales. 
pzmap(sys2)
```
